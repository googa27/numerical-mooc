# Boundary Element Method (BEM)

The BEM is a numerical and computational method to solve linear partial differential equations which have been formulated as integral equations. This integral equation can be taken as an exact solution. The boundary element method attempts to use the given boundary conditions to fit boundary values into the integral equation. Then, the integral equation can be used again to calculate numerically the solution directly at any desired point in the interior of the solution domain. This method is applicable to problems for which Green's functions can be calculated.

## Weak Formulation

The main integral statement of BEM can be interpreted as a combination of a weighted residual statement and a process of integration by parts that reduces or weakens the order of the continuity for the function ($u$).

Let study the Laplace equation, ${\nabla}^{2}{u}$.

#### Weak formulation for Laplace equation:

\begin{equation}
    \int_{\Omega}{\nabla}^{2}{u}\;.\;{w}\;{d\Omega} = 0 \qquad (1)
\end{equation}

Where $w$ is a test function. 

Then, using the second Green's Identity we can write the previous equation as:

\begin{equation}
    0 = \int_{\Omega}{\nabla}^{2}{u}\;.\;{w}\;{d\Omega} = \int_{\Omega}{u}\;.\;{\nabla}^{2}{w}\;{d\Omega} \;+\;
        \int_{\Gamma} \left[ w\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}\;-\; {u}\;.\; \frac{\partial{w}}{\partial{\mathbf{n}}}\right]
        \; {d\Gamma} \qquad (2)
\end{equation}

Where $d\Gamma$ is the boundary of $\Omega$.

If we choose the Green's function as the test function $w$, i.e. $w=G(\textbf{x},\textbf{x'})= \frac{1}{4\pi|\textbf{x}-\textbf{x'}|}$. Then

\begin{equation}
    {\nabla}^{2}{G(\mathbf{x},\mathbf{x'})}= - \delta({\mathbf{x}-\mathbf{x'}}) \qquad (3)
\end{equation}

So equation number $(2)$ can be written as:

\begin{equation}
    0 = \int_{\Omega}{u}\;.\;{\nabla}^{2}{G(\mathbf{x},\mathbf{x'})}\;{d\Omega} \;+\;
        \int_{\Gamma} \left[ G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}\;-\; {u}\;.\;
        \frac{\partial{G}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'})\right]
        \; {d\Gamma} 
\end{equation}

Using (3) we have and taking into account that $u=u(\textbf{x})$:

\begin{equation}
    0 = -\int_{\Omega}{u}\;.\;\;{\delta({\mathbf{x}-\mathbf{x'}})}{d\Omega} \;+\;
        \int_{\Gamma} \left[ G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}\;-\; {u}\;.\;
        \frac{\partial{G}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'})\right]
        \; {d\Gamma}
\end{equation}

Finally,

\begin{equation}
    u(\mathbf{x'}) = \int_{\Gamma} \left[ G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}\;-\; {u}\;.\;
        \frac{\partial{G}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'})\right]
        \; {d\Gamma} \qquad (4)
\end{equation}

The common notation:

   * $\textbf{x}$ for the field point or receiver.
   * $\textbf{x'}$ for the load point or source.

Therefore, when we have a problem based in Laplace equation, we can reformulate it into an integral formulation, then to solve the problem we should solve equation (4). To do this we have to take into account the singularities in $\textbf{x'}$ due to $G(\textbf{x},\textbf{x'})$ and $\frac{\partial{G}}{\partial{\textbf{n}}}(\textbf{x},\textbf{x'})$.

The equation $(4)$ is valid for any point within the $\Omega$ domain. In BE it is usally preferable for computational reasons to apply $(4)$ on the boundary and hence we need to find out what happens when the point $\textbf{x'}$ is on $\Gamma$. A simple way to address this is to consider that $\textbf{x'}$ is in the boundary but the dommain itself is augmented by a hemisphere of radius $\epsilon$. In 3D we have:

Therefore, for the first term in the RHS of equation $(4)$, we have:

\begin{equation}
     \int_{\Gamma_{\epsilon}} G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}(\mathbf{x}) \; {d{\Gamma}_{\epsilon}}
     = \lim_{\epsilon\to\ 0} \left\{ {\int_{{\Gamma}_{\epsilon}} G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}
     (\mathbf{x})\; {d{\Gamma}_{\epsilon}}} \right\} = \lim_{\epsilon\to\ 0} \left\{ {\int_{{\Gamma}_{\epsilon}}
     \frac{\partial{u}}{\partial{\mathbf{n}}}(\mathbf{x})\; . \; \frac{1}{4{\pi}{\epsilon}} {d{\Gamma}_{\epsilon}}} \right\} = 
     \lim_{\epsilon\to\ 0} \left\{\frac{\partial{u}}{\partial{\mathbf{n}}}(\mathbf{x})\; . \; \frac{2{\pi}{\epsilon}^2}{4{\pi}
     {\epsilon}} \right\} = 0 
\end{equation}

and for the second term in the RHS of the of equation $(4)$, we have:

\begin{equation}
     \int_{\Gamma_{\epsilon}}{u}(\mathbf{x}) \;.\; \frac{G(\mathbf{x},\mathbf{x'})}{\partial{\mathbf{n}}} \; {d{\Gamma}_{\epsilon}}
     = \lim_{\epsilon\to\ 0} \left\{ {\int_{{\Gamma}_{\epsilon}} u(\mathbf{x})\;.\; \frac{G(\mathbf{x},\mathbf{x'})}{\partial{\mathbf{n}}}
     \; {d{\Gamma}_{\epsilon}}} \right\} = \lim_{\epsilon\to\ 0} \left\{ {\int_{{\Gamma}_{\epsilon}}
     {-u}(\mathbf{x})\; . \; \frac{1}{4{\pi}{\epsilon}^2} {d{\Gamma}_{\epsilon}}} \right\} = \lim_{\epsilon\to\ 0} \left\{
     {{-u}(\mathbf{x})\; . \; \frac{2{\pi}{\epsilon}^2}{4{\pi}{\epsilon}^2} {d{\Gamma}_{\epsilon}}} \right\} = -\frac{1}{2}{u}(\mathbf{x'})
\end{equation}


Where we use ${d{\Gamma}_{\epsilon}} = {\epsilon}^2 sin(\theta)d{\theta}d{\phi}$, i.e. the diferential of spherical coordinates and $|\textbf{x}-\textbf{x'}|=\epsilon$

Then looking at $(4)$ we can write:

\begin{equation}
       \frac{1}{2}u(\mathbf{x'}) = \int_{\Gamma} \left[ G(\mathbf{x},\mathbf{x'})\;.\; \frac{\partial{u}}{\partial{\mathbf{n}}}(\mathbf{x})
       \;-\; {u}(\mathbf{x})\;.\; \frac{\partial{G}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'})\right]
        \; {d\Gamma} \qquad (5)
\end{equation}

Where all the integrals are in the sense of Cauchy Principal Value. This equation is the boundary integral equation generally used as a starting point for boundary elements.

Let us now consider how expression $(5)$ can be discretized to find a system of equations from which the boundary values can be found. 
Just to figure out we are going to think in a 2D body and we are going to divide the boundary of this body into $N$ segments or elements.

####Constant elements

The values of $u(\textbf{x})$ and $\frac{\partial{u}}{\partial{\textbf{n}}}(\textbf{x})$ are assumed to be constant over each element and equal to the value at the mid-element node.
The points where the unknown values are considered are called "nodes" and taken to be i the middle of the element.

Then the equation $(5)$ can be discretized for a given point "$i$" before applying any boundary condition, as:

\begin{equation}
       \frac{1}{2}u_{i}(\mathbf{x'}) = \sum_{j=1}^{N} \frac{\partial{u_{j}}}{\partial{\mathbf{n}}}(\mathbf{x}) \; \int_{\Gamma_{j}}
       G(\mathbf{x},\mathbf{x'}) \; {d\Gamma_j}\;-\; \sum_{j=1}^{N} {u_j}(\mathbf{x}) \; \int_{\Gamma_{j}}
       \frac{\partial{G}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'})\; {d\Gamma_j} \qquad (6)       
\end{equation}

Sometimes the integral terms are called influence coefficients. These integrals relate the "$i$" node, where the fundamental solution is acting, to any other "$j$" node.


If we call:

\begin{equation}
    {\mathbf{G}}_{ij}= \int_{\Gamma_{j}}G_{i}(\mathbf{x},\mathbf{x'}) \; {d\Gamma_j} \qquad \text{and} \qquad 
    {\mathbf{\widetilde{G}}}_{ij}= \int_{\Gamma_{j}}\frac{\partial{G_i}}{\partial{\mathbf{n}}}(\mathbf{x},\mathbf{x'}) \; {d\Gamma_j}
\end{equation}

where the fundamental solution is applied to a particular node $"i"$. Then the equation $(6)$ can be written as:

\begin{equation}
    \frac{1}{2}u_i(\mathbf{x'}) = \sum_{j=1}^{N} {\mathbf{G}}_{ij} \frac{\partial{u_{j}}}{\partial{\mathbf{n}}}(\mathbf{x}) \; - \;
    \sum_{j=1}^{N} {\mathbf{\widetilde{G}}}_{ij} u_j(\mathbf{x})
\end{equation}

If we now assume that the possition od $"i"$ can also vary from $1$ to $N$, i.e. we assume the fundamental solution is applied at each node successively, one obtains a system of equations resulting from applying $(6)$ to each boundary point in turn.

Let us now call:

\begin{equation}
    {\mathbf{\overline{G}}}_{ij}=
\begin{cases}
    {\mathbf{\widetilde{G}}}_{ij} & \text{, } \; {i}\neq{j} \\
    {\mathbf{\widetilde{G}}}_{ij}\;+\; \frac{1}{2}  & \text{, } \; i=j
\end{cases}
\end{equation}

Then for the whole problem, we have the following system of equations:

\begin{equation}
    \sum_{i,j=1}^{N} {\mathbf{\overline{G}}}_{ij}\;{u_j} =\sum_{i,j=1}^{N} {\mathbf{{G}}}_{ij} \frac{\partial{u_{j}}}{\partial{\mathbf{n}}}
\end{equation}

The system looks like:

\begin{equation}
\begin{bmatrix}
  {\mathbf{\widetilde{G}}}_{11} + 1/2 & {\mathbf{\widetilde{G}}}_{12} &\cdots & {\mathbf{\widetilde{G}}}_{1N} \\
  {\mathbf{\widetilde{G}}}_{21} & {\mathbf{\widetilde{G}}}_{22}+ 1/2 &\cdots & {\mathbf{\widetilde{G}}}_{2N} \\
  \vdots & \vdots &\ddots &  \vdots \\
   {\mathbf{\widetilde{G}}}_{N1} & \cdots & & {\mathbf{\widetilde{G}}}_{NN}\;+\; 1/2
\end{bmatrix} 
\begin{bmatrix}
  u_1  \\ u_2 \\ \vdots \\ \\  u_N
\end{bmatrix}
=
\begin{bmatrix}
  {\mathbf{\tilde{G}}}_{11} & {\mathbf{\tilde{G}}}_{12} &\cdots & {\mathbf{\tilde{G}}}_{1N} \\
  {\mathbf{\tilde{G}}}_{21} & {\mathbf{\tilde{G}}}_{22} &\cdots & {\mathbf{\tilde{G}}}_{2N} \\
  \vdots & \vdots &\ddots &  \vdots \\
   {\mathbf{\tilde{G}}}_{N1} & \cdots & & {\mathbf{\tilde{G}}}_{NN}
\end{bmatrix} 
\begin{bmatrix}
  \frac{\partial{u_1}}{\partial{\mathbf{n}}} \\ \vdots \\ \frac{\partial{u_N}}{\partial{\mathbf{n}}}
\end{bmatrix}
\end{equation}

####Collocation Method:

The principle of collocation means to locate the load point $(\textbf{x'})$ sequentially at all nodes of the discretization such that the domain variable at the load point $(u(\textbf{x'}))$ coincides with the nodal value.
